# 🏥 Fase 2: Análisis Exploratorio y Limpieza de Datos

Este notebook contiene el análisis inicial del dataset **Diabetes 130-US Hospitals (1999-2008)**.
Se centra en la comprensión de las variables críticas, la detección de valores faltantes (identificados como `?`) y la preparación para modelos generativos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set_style("whitegrid")

%matplotlib inline

print("✅ Entorno listo")
```

## 1. Carga del Dataset
Cargamos los datos desde la carpeta compartida.

```python
path_data = "../data/diabetic_data.csv"
df = pd.read_csv(path_data)

print(f"Dimensiones originales: {df.shape}")
df.head()
```

## 2. Análisis de Valores Faltantes
En este dataset, los valores nulos están representados por `?`.

```python
# Reemplazamos '?' por NaN para análisis estándar
df_nan = df.replace('?', np.nan)

missing_values = df_nan.isnull().sum()
missing_percent = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({'Nulos': missing_values, 'Porcentaje': missing_percent})
missing_df = missing_df[missing_df['Nulos'] > 0].sort_values(by='Porcentaje', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=missing_df.index, y=missing_df['Porcentaje'])
plt.title("Porcentaje de Valores Faltantes por Columna")
plt.xticks(rotation=45)
plt.show()

missing_df
```

## 3. Distribución de Variables Clave
Analizamos la variable objetivo `readmitted` y variables demográficas.

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Variable Objetivo
sns.countplot(x='readmitted', data=df, ax=axes[0], palette='viridis')
axes[0].set_title("Distribución de la Variable Objetivo (Readmitted)")

# Edad
sns.countplot(y='age', data=df, ax=axes[1], palette='magma')
axes[1].set_title("Distribución por Rangos de Edad")

plt.tight_layout()
plt.show()
```
